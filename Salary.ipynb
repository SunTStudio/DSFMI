<h3>1.2.	Identifikasi Permasalahan Bisnis<h3>

Dalam analisis pengaruh pengalaman dan umur terhadap tinggi gaji pekerja, ada beberapa permasalahan bisnis yang perlu diidentifikasi agar hasil analisis menjadi lebih relevan dan akurat. Beberapa permasalahan tersebut antara lain:
1.	Permasalahan Data yang Tidak Lengkap atau Tidak Akurat: Data yang tidak lengkap atau tidak akurat dapat mengakibatkan distorsi dalam hasil analisis. Misalnya, jika terdapat kekosongan data pengalaman kerja atau usia pekerja, analisis menjadi kurang representatif dan mungkin tidak dapat memberikan kesimpulan yang valid.
2.	Permasalahan Penyimpangan Data (Outliers): Adanya data yang jauh berbeda dari sebagian besar data lainnya dapat menyebabkan penyimpangan dalam analisis. Outliers ini bisa mempengaruhi hasil analisis secara signifikan dan harus ditangani dengan cermat.
3.	Korelasi Tidak Menunjukkan Kausalitas: Meskipun analisis mungkin menemukan hubungan antara pengalaman kerja, usia, dan tinggi gaji, perlu diingat bahwa korelasi tidak selalu menunjukkan adanya hubungan sebab-akibat. Ada kemungkinan terdapat faktor lain yang tidak diidentifikasi dalam analisis yang mempengaruhi tinggi gaji pekerja.
4.	Pengaruh Faktor Lain yang Tidak Diukur: Dalam analisis ini, fokus pada pengalaman kerja dan usia sebagai faktor yang mempengaruhi gaji. Namun, terdapat kemungkinan bahwa ada faktor lain yang tidak diukur dalam analisis yang juga memiliki pengaruh signifikan terhadap tinggi gaji, seperti tingkat pendidikan, tingkat tanggung jawab pekerjaan, atau jenis industri tempat pekerja bekerja.
5.	Masalah Multikolinearitas: Multikolinearitas adalah fenomena ketika dua atau lebih variabel independen saling berkorelasi kuat. Hal ini bisa menyebabkan kesulitan dalam mengisolasi efek dari setiap variabel independen secara individu dan menyulitkan interpretasi hasil analisis.
6.	Perbedaan dalam Struktur Gaji: Setiap perusahaan atau industri mungkin memiliki struktur gaji yang berbeda. Pengaruh pengalaman dan usia terhadap tinggi gaji dapat bervariasi tergantung pada bagaimana perusahaan atau industri tersebut menentukan kebijakan penggajian.
7.	Perubahan Lingkungan Bisnis: Analisis ini mungkin tidak mempertimbangkan perubahan lingkungan bisnis yang dapat memengaruhi tinggi gaji pekerja dari waktu ke waktu.
8.	Aspek Hukum dan Etika: Penggunaan data terkait usia pekerja harus memperhatikan aspek hukum dan etika yang berlaku terkait perlindungan data dan kebijakan anti-diskriminasi.


<h3>1.3.	Asumsi, Kebutuhan, dan Limitasi<h3>

Dalam melakukan analisis pengaruh pengalaman dan umur terhadap tinggi gaji pekerja, terdapat beberapa asumsi, kebutuhan, dan limitasi yang perlu diperhatikan:
Asumsi:
1.	Linearitas hubungan: Analisis ini didasarkan pada asumsi bahwa hubungan antara pengalaman, umur, dan tinggi gaji adalah linier. Artinya, diasumsikan bahwa setiap peningkatan unit dalam pengalaman atau umur akan memiliki dampak yang konsisten pada tinggi gaji.
2.	Independensi data: Asumsi ini mengandaikan bahwa setiap data dalam analisis adalah independen satu sama lain. Dalam konteks ini, berarti bahwa data gaji, pengalaman, dan umur pekerja tidak dipengaruhi oleh data pekerja lainnya.
3.	Normalitas data: Analisis statistik yang digunakan, seperti regresi linier atau ANOVA, mengasumsikan bahwa data mengikuti distribusi normal. Jika data tidak memenuhi asumsi normalitas, perlu dilakukan transformasi data atau menggunakan metode analisis lain yang sesuai.
Kebutuhan:
1.	Data yang tepat: Untuk melakukan analisis ini, diperlukan data yang lengkap dan tepat mengenai gaji, pengalaman kerja, dan usia pekerja. Data yang tidak akurat atau tidak lengkap dapat mengurangi validitas dan reliabilitas hasil analisis.
2.	Metode analisis yang sesuai: Pemilihan metode analisis yang tepat sangat penting. Regresi linier berguna untuk menilai hubungan antara variabel dependen (tinggi gaji) dengan satu atau lebih variabel independen (pengalaman dan umur). Namun, tergantung pada kompleksitas data dan tujuan analisis, mungkin juga perlu mempertimbangkan metode analisis lain seperti analisis multivariat atau analisis kelompok.
3.	Pengelompokan pekerjaan: Dalam analisis ini, penting untuk mempertimbangkan bahwa berbagai jenis pekerjaan atau industri mungkin memiliki pola hubungan yang berbeda antara pengalaman, umur, dan tinggi gaji. Oleh karena itu, pengelompokan pekerjaan atau industri dapat membantu dalam analisis yang lebih tepat.
Limitasi:
1.	Korelasi bukan kausalitas: Meskipun analisis ini dapat menunjukkan hubungan antara pengalaman dan umur dengan tinggi gaji, penting untuk diingat bahwa korelasi tidak selalu menyiratkan hubungan sebab-akibat. Ada kemungkinan bahwa faktor-faktor lain, seperti pendidikan, lokasi geografis, atau kebijakan perusahaan, juga turut berkontribusi terhadap tinggi gaji pekerja.
2.	Pengaruh variabel lain: Analisis ini mungkin tidak mempertimbangkan secara lengkap semua faktor yang dapat mempengaruhi tinggi gaji. Variabel-variabel seperti pendidikan, jenis pekerjaan, jenis industri, atau tingkat ekonomi negara dapat memainkan peran penting dalam menentukan tinggi gaji pekerja.
3.	Data longitudinal: Analisis ini mungkin dilakukan dengan data lintas-seksi atau lintas-waktu. Namun, analisis dengan data longitudinal (data yang dikumpulkan dari subjek yang sama dalam beberapa titik waktu) dapat memberikan wawasan yang lebih dalam mengenai perubahan gaji seiring bertambahnya pengalaman dan umur pekerja.
4.	Perubahan seiring waktu: Analisis ini mungkin tidak mempertimbangkan perubahan kebijakan atau kondisi pasar kerja yang terjadi seiring waktu. Faktor-faktor ini dapat mempengaruhi hubungan antara pengalaman, umur, dan tinggi gaji pekerja dari tahun ke tahun.

<h2>4.	Instrumen Pengukuran Keberhasilan <h2>

Untuk mengukur keberhasilan dari analisis pengaruh pengalaman dan umur terhadap tinggi gaji pekerja, beberapa instrumen pengukuran yang dapat digunakan antara lain:
1.	Koefisien Korelasi: Instrumen ini dapat digunakan untuk mengukur sejauh mana ada hubungan antara pengalaman kerja, usia, dan tinggi gaji pekerja. Koefisien korelasi akan memberikan nilai antara -1 hingga 1, di mana nilai positif menunjukkan hubungan positif, nilai negatif menunjukkan hubungan negatif, dan nilai mendekati nol menunjukkan tidak adanya hubungan.
2.	Analisis Regresi Linier: Instrumen ini membantu mengukur pengaruh variabel independen (seperti pengalaman kerja dan usia) terhadap variabel dependen (tinggi gaji pekerja). Regresi linier akan menghasilkan persamaan garis regresi yang menunjukkan seberapa besar perubahan tinggi gaji yang diharapkan terjadi ketika pengalaman kerja atau usia berubah.
3.	Uji Signifikansi Statistik: Instrumen ini berguna untuk mengukur signifikansi statistik dari hubungan antara pengalaman kerja dan usia terhadap tinggi gaji pekerja. Dengan menggunakan uji statistik seperti uji t atau uji F (untuk lebih dari dua kelompok umur), kita dapat menentukan apakah hubungan tersebut secara signifikan berpengaruh terhadap tinggi gaji pekerja.
4.	Analisis Varians (ANOVA): Jika terdapat lebih dari dua kelompok umur, ANOVA dapat digunakan untuk membandingkan rata-rata gaji antar kelompok umur dan menentukan apakah perbedaan tersebut signifikan.
5.	Analisis Perbedaan Rata-rata: Instrumen ini dapat digunakan untuk membandingkan rata-rata tinggi gaji pekerja antara kelompok yang berbeda berdasarkan pengalaman kerja (misalnya, pemula, berpengalaman, dan veteran) atau kelompok umur (kelompok usia 20-30 tahun, 31-40 tahun, dan seterusnya).
6.	Persentil Gaji: Mengukur tinggi gaji pada persentil tertentu (misalnya, gaji 25 persen terendah, median, atau gaji 75 persen tertinggi) dapat memberikan gambaran yang lebih rinci tentang distribusi tinggi gaji dalam kelompok pengalaman atau umur tertentu.
7.	Analisis Kepuasan Karyawan: Selain angka-angka statistik, pendekatan kualitatif juga penting. Melakukan wawancara atau survei untuk mengetahui bagaimana karyawan merasa tentang sistem penggajian dan bagaimana mereka menilai keadilan gaji berdasarkan pengalaman dan umur dapat memberikan wawasan tambahan.
Penting untuk mencatat bahwa dalam analisis ini, tidak hanya faktor pengalaman dan umur yang harus dipertimbangkan. Faktor-faktor lain seperti tingkat pendidikan, industri, dan peran pekerja juga dapat mempengaruhi tinggi gaji pekerja. Oleh karena itu, analisis ini sebaiknya dilakukan secara komprehensif untuk mendapatkan pemahaman yang lebih lengkap tentang faktor-faktor yang berkontribusi pada tinggi gaji pekerja.

```python
# the code below is adopted from this https://github.com/ashishpatel26/Amazing-Feature-Engineering/blob/master/feature_cleaning/missing_data.py

def check_missing(data,output_path=None):
    """
    check the total number & percentage of missing values
    per variable of a pandas Dataframe
    """

    result = pd.concat([data.isnull().sum(),data.isnull().mean()],axis=1)
    result = result.rename(index=str,columns={0:'total missing',1:'proportion'})
    if output_path is not None:
        result.to_csv(output_path+'missing.csv')
        print('result saved at', output_path, 'missing.csv')
    return result
```

```python
#3. masukkan kolom kedalam varibel
X = df['Years of Experience'].values.reshape(-1,1) #kolom pengalaman
y = df['Salary'].values.reshape(-1,1) #kolom gaji
```

```python
#4. split data u/ training 70% dan u/ testing 30%
from sklearn.model_selection import train_test_split

x_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
#5. analisis regresi
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#6. cek akurasi
round(model.score(X_test,y_test) * 100 ,2)
```

```python
#7. buat visualisasinya
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize=(10,5))

ax.scatter(x_train,y_train,c='b')
ax.plot(x_train,model.predict(x_train),c='r')
ax.scatter(X_test,y_test,c='y')

ax.set_xlabel('Pengalaman')
ax.set_ylabel('Gaji')
ax.set_title('Grafik Pengalaman dan Gaji')

plt.show()
```
